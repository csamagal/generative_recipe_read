```python
bucket_name = "cs-trial-s3-storage"

s3 = boto3.client("s3")
response = s3.list_objects_v2(Bucket=bucket_name)

for obj in response.get("Contents", []):
    print(obj["Key"])
```

```python
old_prefix = "drive-download-20260202T202227Z-3-001/"
new_prefix = "recipe_jpgs/"

s3 = boto3.resource("s3")
bucket = s3.Bucket(bucket_name)

for obj in bucket.objects.filter(Prefix=old_prefix):
    old_key = obj.key
    new_key = old_key.replace(old_prefix, new_prefix, 1)

    print(f"Moving {old_key} → {new_key}")

    # Copy to new location
    s3.Object(bucket_name, new_key).copy_from(
        CopySource={"Bucket": bucket_name, "Key": old_key}
    )

    # Delete old object
    s3.Object(bucket_name, old_key).delete()

print("✅ Folder rename complete")
```
